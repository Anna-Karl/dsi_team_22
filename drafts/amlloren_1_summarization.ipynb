## 2. Understanding the Schema

#### 1. Percentage of Health Expenditure (PC_HEALTHXP)
**Definition:** This is the percentage of a country's total health expenditure specifically spent on pharmaceuticals, including prescription medicines and over-the-counter products.

**Explanation:** Imagine a country’s health budget like a pie. This number tells us what slice of the pie is spent on medicines, including both prescription drugs and over-the-counter products. For example, if a country spends 15% of its health budget on pharmaceuticals, it means that 15 cents of every dollar allocated to health services are used for medicines.

#### 2. Percentage of GDP (PC_GDP)
**Definition:** This is the percentage of the country's Gross Domestic Product (GDP) that is spent on pharmaceuticals.

**Explanation:** GDP is like the total income of a country. This percentage tells us how big the pharmaceutical spending is compared to the entire economy. For example, if the percentage is 0.7%, that means out of the entire economic output of the country, 0.7% is spent on medicines. It highlights the significance of pharmaceutical expenditures within the broader economic context.

#### 3. Health Expenditure per Capita in USD (USD_CAP)
**Definition:** This is the average amount of money spent on pharmaceuticals per person, calculated in US dollars.

**Explanation:** Per capita means per person. This number tells us how much, on average, each person in the country spends on medicines in a year. If this number is $500, it means that on average, each person in the country has $500 worth of pharmaceuticals. This helps understand the financial burden or investment in medicines for the average citizen.

#### 4. Total Spending (TOTAL_SPEND)
**Definition:** This is the total amount of money spent on pharmaceuticals by the country in a given year, typically in millions or billions of dollars.

**Explanation:** This number gives us the overall scale of pharmaceutical expenditure in the country. It’s like looking at the total bill for all medicines consumed in the country for the year. If the total spending is $50 billion, that’s the sum of all money spent on prescription and over-the-counter drugs, excluding those consumed in hospitals and other health care settings.

## 3. What These Numbers Tell Us:

* __Comparing Countries:__ By looking at these numbers, we can compare how different countries prioritize and invest in pharmaceuticals. For example, a country with a high percentage of pharmaceutical expenditure might rely more on medications for health care compared to a country with a lower percentage.

* __Trends Over Time:__ Observing these numbers over different years can show us trends, such as whether a country is increasing or decreasing its investment in pharmaceuticals.

* __Economic Context:__ The percentage of GDP spent on pharmaceuticals helps understand pharmaceutical spending in relation to the country's overall economy.

* __Individual Impact:__ Pharmaceutical expenditure per capita gives insight into how much financial resources are allocated to each individual for medicines.

#### Practical Example:

This can help stakeholders understand and make informed decisions about health care policies and funding.

### Visualization: Country Observations by Year

To visualize which years each country has an observation, we can use a heatmap. A heatmap is a great way to represent the presence or absence of observations across different years for each country. We can use the seaborn and matplotlib libraries in Python to create this visualization.

Step-by-Step Guide
* __Prepare the Data:__ Transform the data into a format suitable for a heatmap.
* __Create the Heatmap:__ Use seaborn to create the heatmap.

The resulting heatmap will have countries on the y-axis (ordered by the number of observations from highest to lowest) and years on the x-axis. Cells will be colored based on the presence (1) or absence (0) of observations, providing a clear visualization of which years each country has data available.

This approach ensures that countries with the most observations are displayed at the top, making it easier to identify data-rich countries. Adjusting the figsize parameter ensures all labels are readable.

#### Supporting Python Code:

```python
# Create a pivot table with countries as rows and years as columns
pivot_table = df.pivot_table(index='LOCATION', columns='TIME', aggfunc='size', fill_value=0)

# Convert the pivot table to a binary format (presence or absence of observation)
pivot_table = pivot_table.applymap(lambda x: 1 if x > 0 else 0)

# Count the number of observations per country
observation_counts = pivot_table.sum(axis=1)

# Sort the countries by the number of observations
sorted_countries = observation_counts.sort_values(ascending=False).index

# Reorder the pivot table based on the sorted countries
pivot_table_sorted = pivot_table.loc[sorted_countries]

# Plot the heatmap
plt.figure(figsize=(16, 10))
heatmap = sns.heatmap(pivot_table_sorted, cmap="YlGnBu", cbar=False, linewidths=.5)

# Add labels and title
heatmap.set_xlabel('Year', fontsize=14)
heatmap.set_ylabel('Country', fontsize=14)
heatmap.set_title('Observations of Each Country by Year', fontsize=18)

# Display the heatmap
plt.show()
```
